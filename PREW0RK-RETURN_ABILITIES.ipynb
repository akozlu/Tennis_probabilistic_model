```python
x=frame_2['winner_name'].unique().tolist()
v=frame_2['loser_name'].unique().tolist()
res_f=list(set(x+v))
len(res_f)
```

```python
x=frame['Player_1'].unique().tolist()
y=frame['Player_2'].unique().tolist()

res = list(set(x+y))
len(y),len(x),len(res)

final_list=list(set(res_f and res))
len(final_list)
```

```python
def win_perc_total(df,final_list):
    final=[]
    
    final_list=sorted(final_list)
    
    for z in range(len(sorted(final_list))):
        points=[]
        points.append(final_list[z])
        df_11=df[(df['row']=='v1st') & (df['Player_1']==final_list[z]) & (df['player']==1)]
        t_1=df_11['pts'].sum()
        points.append(t_1)
        
        r_1=df_11['returnable'].sum()
        points.append(r_1)
        
        r_w_1=df_11['returnable_won'].sum()
        points.append(r_w_1)
        
        df_12=df[(df['row']=='v2nd') & (df['Player_1']==final_list[z]) & (df['player']==1)]
        t_2=df_12['pts'].sum()
        points.append(t_2)
        
        r_2=df_12['returnable'].sum()
        points.append(r_2)
        
        r_w_2=df_12['returnable_won'].sum()
        points.append(r_w_2)
        
        final.append(points)
        
    return final
            
                   
                   
                   
    
```

```python
t=win_perc_total(frame,final_list)
t
```

```python
columns=['Player_name','T_1','R_1','R_1w','T_2','R_2','R_2w']
df_ret=pd.DataFrame(t,columns=columns)
```

```python
df_ret=df_ret.dropna()
df_ret
```

```python
df_ret.to_csv('return_analysis.csv', encoding='utf-8', index=False)
```
